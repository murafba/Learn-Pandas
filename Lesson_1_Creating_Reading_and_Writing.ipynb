```python
import pandas as pd
```

## DataFrame

```python
pd.DataFrame({'Yes' : [50, 21], 'No' : [131, 2]})
```

```python
# Using row labels
pd.DataFrame({'Bob': ['I like it.', 'It was awful.'],
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

## Series
*Series* is a sequence of data values. A series is a list.

```python
pd.Series([1,2,3,4,'5'], index=['a', 'b', 'c', 'd', 'e'])
```

```python
# A series doens not have a column name. It only has one overall [name]
# Use [index] params to meke row labels
pd.Series([20, 30], index=['2015 sales', '2016 sales'], name='Product A')
```
