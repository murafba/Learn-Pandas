## Constitional Selection

```python
reviews.country == 'Italy'
```

```python
# The result above can ben used inside of [loc] to select the relevant data
reviews.loc[reviews['country'] == 'Italy']
```

```python
reviews.loc[(reviews['country'] == 'Italy') & (reviews['points'] >= 90)]
```

```python
# Suppose we would like to buy wine that's made in Italy or which is rated above average
reviews.loc[(reviews['country'] == 'Italy') | (reviews['points'] >= 90)]
```

```python
### Two built-in selectors in Pandas
reviews.loc[reviews.country.isin(['Italy'])]
```

```python
reviews.loc[reviews.price.notnull()]
```

```python
reviews.loc[reviews.price.isnull()]
```
